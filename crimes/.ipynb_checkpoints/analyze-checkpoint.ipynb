```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv("crimes_new.csv", parse_dates=['date'])
# data.loc[:, data.columns.isin(['grid', 'beat', 'district', 'address'])].head()
```

```python
fig, ax = plt.subplots(figsize=(14,5))
data.groupby(['date']).count()['description'].plot(ax=ax, title='Crimes committed, January 2006 Sacramento', style='o-')
plt.show()
```

```python
gb_code = data.groupby(['code_description'])
gb_code_sorted = gb_code.count().sort_values('date', ascending=False)

print('Top 10 most popular crime types')
gb_code_sorted.loc[:, ['date']][1:11]
```
